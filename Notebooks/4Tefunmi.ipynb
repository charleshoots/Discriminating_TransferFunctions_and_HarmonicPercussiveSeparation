```python
import numpy as np
from obspy import Catalog
from obspy.clients.fdsn import Client
from obspy import UTCDateTime

minmag=6.5
# start = '2023-05-10T16:01:59'
start = '2023-01-01' #Start of Galapagos deployment
start = UTCDateTime(start)
end = start + 18*30*24*3600 #end of deployment
current_start = start
interval_width_sec = 24*30*3600 #Search interval width (in seconds)
client = Client();catalog_list=Catalog()
while current_start<end:
    current_end = current_start+interval_width_sec
    catalog_list+=client.get_events(starttime=current_start, endtime=current_end,minmagnitude=minmag,orderby='magnitude')
    current_start+=interval_width_sec

# Sorts final list by magnitude
catalog_list=Catalog([catalog_list[i] for i in np.flip(np.argsort(([c.magnitudes[0].mag for c in catalog_list])))])
```
