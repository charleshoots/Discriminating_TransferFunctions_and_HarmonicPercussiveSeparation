# Quality control procedure

Developing a high-quality dataset that accurately and reliably captures the variability in noise spectral characteristics across major OBS deployments required meticulous quality control and iterative refinement. Several key factors were assessed to ensure data integrity, including overall noise levels, data continuity, and the identification of consistently aberrant stations. Additionally, criteria were established to maintain sufficient station density per event, a minimum number of recorded events per station, and an adequate duration of noise observations for each station. These measures were implemented through an iterative process of data assessment and refinement of compromised data. These two stages are repeated in sequence until a final data that had satisfied our quality controls was achieved.

The quality control process followed two main stages:
## | STAGE ONE: DATA ASSESSMENT
This initial phase focused on identifying and mitigating issues that could compromise the dataset’s reliability, particularly data gaps, noise anomalies, and consistently aberrant stations. The following tools are used in this stage to evaluate the data set quality controls listed below:

                1. Day and station averaged noise coherence and phase spectra for cross-component data ZP, Z1, Z2, and ZZ (pre/post correction) below 1hz.
                2. Day and station averaged noised power spectral density below 1hz.
                3. Station, network, depth and instrument averaged event ZZ (pre/post correction) coherences below 1hz.
                4. Statistical evaluation of deployment and event meta-parameters using stem-plots and histograms.
                5. Available reports of day and station averaged spectra from Mustang


    Data Gaps:
Inconsistent data availability was detected at multiple stations, with gaps that were not reflected in the IRIS-DMC repository. Approximately 2.6% of the initial dataset contained missing fragments, which, if left unaddressed, could lead to errors in noise and transfer function estimates. Gaps occurring within noise traces could distort spectral estimates, while those within event traces could corrupt data retention when processed using NoiseCut or ATaCR. Despite being a rare occurrence, the statistical impact of these gaps was significant, requiring careful management.

To detect these gaps, an initial heuristic script was employed, scanning for indicators such as flat amplitudes (zeros), NaNs, and empty traces. This method proved highly effective, identifying the majority of data gaps.
Subsequently, a manual review was conducted, where an analyst examined more than 10,000 event traces and noise spectra, ensuring no overlooked issues remained.

    Noise Quality:
During the manual review process, analysts also evaluated the spectral characteristics of all four components at each station (pressure, vertical, and two horizontal channels). Data exhibiting the following anomalies were flagged and removed from the dataset:

        i. Mid-deployment gain changes affecting noise spectra.
        i. Significant electrical noise contamination.
        i. Inconsistent spectral densities when compared to IRIS Mustang reports of daily average spectrograms for the corresponding station.
        i. A complete absence of microseismic peaks, which suggested faulty instrumentation or recording errors.

Some stations exhibited persistent anomalies that rendered them unsuitable for this study. Examples of consistently aberrant station included:
        ii. Electrical noise contamination, such as that observed at W06 in the Keck deployment.
        ii. Persistent standing sine waves with no apparent response to ambient noise conditions, as observed at ENAM’s X09 deployed BBOBS.
        Stations exhibiting these behaviors were systematically removed from the dataset to maintain integrity.

## | STAGE TWO: DATA REFINEMENT:

    Maintaining Observational Volume
To preserve dataset completeness and statistical robustness, the above data quality assessments were repeated following any modification. WHen a day of noise or station was permanently removed, it was replaced with a new noise sample or an additional station with similar deployment parameters. 15 days of noise is initially collected for each station to provide a 5-day cushion for the target minimum of 10-days of quality noise spectra for use in ATaCR.

    Minimum Station Density per Event:
When culling consistently bad stations from the data, we took care to mitigate the loss in density of stations for any affected event with event observations on station(s) used to replace the disfunctional BBOBS. Every event in the catalog was required to maintain observations from atleast 10 different BBOBS.

    Minimum Events per Station:
Each station needed to record a minimum number of seismic events to be retained in the dataset. Stations that failed to meet this requirement were substituted with alternative deployments.

    Minimum Duration per Station:
Stations were required to meet a minimum number of operational days to ensure statistical reliability. Those falling short were supplemented with additional data or alternative stations.

This rigorous and iterative quality control process ensured that the final dataset was both representative and reliable, providing a robust foundation for further analysis of noise variability in major OBS deployments.

## Assessment example: Data gaps
___
Of the two methods, NoiseCut was usually the first to demonstrate an issue with gaps in data since it does not relying on averaged samples of noise for corrections and is thus more susceptible to isolated and acute gaps in data. If there is a gap at any point in the 24hr trace that NoiseCut uses for an event the issue will be apparent through issues such as Gibb's artifacts unsuprisingly even when a gap is outside the 2-hour window the trace is trimmed to after correction. Below is an example of this in Cascadia's J11B were data gaps were present both within and outside the event window (red lines).
___

___
