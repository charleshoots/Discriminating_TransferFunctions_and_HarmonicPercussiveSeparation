
## Pre-AGU QC:
___
Data Quality Control Stage 1 | Noise power spectral densities (PSD) on day and station averages were examined|
----|----|
Reports from Mustang | https://www.zotero.org/groups/5749286/atacr.and.noisecut.comparison/collections/W9U467IU|
Station Average Noise PSDs (updated since AGU)| https://www.zotero.org/groups/5749286/atacr.and.noisecut.comparison/collections/PWSRPHXB|

- Examining the Noise average PSDs for each station and comparing with IRIS's Mustang as a quality check against our own pre-processing and plotting method to see that the variability in noise, at the dB we showed in PSDs, matched station PSDs reported on Mustang.

- Primarily, this quality control served to verify the data and units used (m/s and Pascals) were correctly deconvolved during instrument response removal.

- Using Mustang also provided an initial benchmark for when and where unsual behavior in noise PSD for a station occured and subsequently the days to remove.

- While we knew Mustang was not going to provide the same quality control for pressure data, atleast a third of the dataset still reports on Mustang for pressure data. It is possible more pressure data through Mustang may be available if the API is scraped correctly. This was possible because the online interface does not give options for pressure while direct calls to it's API bypasses this.
___
Data Quality Control Stage 2 | Correction coherence versus deployment parameter variability with respect to other channels (ZZ,ZP,Z1,Z2)|
----|----|
Results and summary report located at| https://www.zotero.org/groups/5749286/atacr.and.noisecut.comparison/collections/GTIV4AV6 |

- Using stem-plots, variability in post correction coherences were examined with respect to station, network, instrument design, depth, and seismometer averages. 

- This highlighted several stations with aberrant channels. The stations were either removed from the catalog entirely or new days of noise were collected to replace unsual ones (ie gain change or inconsistently noisy channels).
___
Data Quality Control Stage 3 | Examined each trace by eye, looking for aberant event traces and coherence averages that indicate station issues|
----|----|
Results and summary report located at| https://www.zotero.org/groups/5749286/atacr.and.noisecut.comparison/collections/MQSLMPGJ |

- Event record sections showing both trace in the time domain and spectral coherence were then examined to control for event variability when looking for unsual station behaviors.

- In addition to showing the same stations with unusual behavior that were indicated by the stem plots, this was also the first time that illustrated the larger event variability in terms of NoiseCut ZZ coherences despite consistently less change (higher ZZ coherence) than ATaCR on average.

- While the ZZ coherence for NoiseCut was generally higher, questions still remain as to what, if any, utility can have been made from event ZP coherences which did not show the same behavior and in many cases increased ZP coherence.
___

Data Quality Control Stage 4 | Examined every single individual noise and event trace, for both ATaCR and NoiseCut by eye looking for gaps|
----|----|
Results and summary reports | https://www.zotero.org/groups/5749286/atacr.and.noisecut.comparison/collections/2GZ64XNH |
Every station with a data gap | https://www.zotero.org/groups/5749286/atacr.and.noisecut.comparison/collections/YH3BTDQM/items/KNW5NZ68/reader|

- More than anything, this was a necessary hands-on quality control step that allowed us to highlight any dead traces since there is no automated quality control for the event traces built into either method. This resulted, in particular for NoiseCut which requires longer traces, inconsistently dead traces on a few stations for which their inclusion in that respective event was removed.

- As the last stage in the quality control before AGU, particular attention is given to looking for unusual station behaviors across all channels using both spectral coherence and phase.

- This stage flagged 17 stations for possible removal, 6 of which were removed and the remaining were fixed new days used for the noise spectra estimate.
___
Data Quality Control Stage 5 (post-AGU) | Adressing record station density, event magnitude distributions, and shallow stations |
----|----|
Notebook on updates to dataset|https://app.reviewnb.com/charleshoots/ATACR_HPS_Comp/blob/main/RunningNotebooks/01.21.25.ipynb/|

- Following AGU, the station density for events to be atleast 10 stations for each. 

- Prior to this, the magnitude distribution was not following the epexcted GR distribution. The fact that it wasn't already is most likely a result of the highly discontinuous and occasionally overlapping deployment windows that were searched for events with no explicit enforcement on minimum station observation density.

- Additionally, the contemporaneous deployment windows were further aided when the AACSE network was entirely removed from the dataset as it did not overlap with any other deployment in the dataset.

- Following the enforced station minimum and removing AACSE, the event magnitude distribution become a more typical GR distribution.
- In total stations, 4 stations (AACSE) were removed and 18 were added. 8 from ENAM and 10 shallow from Cascadia.

- More than 100 <Mw6.7 events were removed for low station density and replaced with around 150 Mw 6-8.

- 14 events were for reporting a magnitude type other than MW. 

- In total events, the catalog was increased by ~30% from 289 prior to AGU to the current 359.
